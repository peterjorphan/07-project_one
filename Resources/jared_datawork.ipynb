```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
file = "../output/weather.csv"
df = pd.read_csv(file)
df
```

```python
# df['New Start'] = df['Trip Start Timestamp'].str.split(pat=None, expand=True, n=-1)
# df['Trip Start Timestamp'].str.split(' ', -1, expand=True)
df[['new time', 'time']] = df['time'].str.split('T', n=1, expand=True)
df
```

```python
# df1 = df.groupby(['summary', 'new time'])['Trip ID'].count()
df1 = df.groupby(['summary', 'new time'], as_index=False).agg({'Trip ID': 'count', 'precipIntensity': 'mean', 'precipProbability': 'mean', 'temperature': 'mean', 'apparentTemperature': 'mean', 'windSpeed': 'mean'})
# df1.to_csv('summary_df.csv')
df1.head()
```

```python
df2 = df1.groupby(['summary'], as_index=False).agg({'Trip ID': 'mean', 'precipIntensity': 'mean', 'precipProbability': 'mean', 'temperature': 'mean', 'apparentTemperature': 'mean', 'windSpeed': 'mean'})
df2
```

```python
df2.to_csv('summary_mean_totals.csv')
```
