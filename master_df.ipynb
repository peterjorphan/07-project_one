```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
file = "./output/weather.csv"
wdf = pd.read_csv(file)
wdf.head()
```

```python
file = "./resources/DataWithHeaders.csv"
tdf = pd.read_csv(file)
tdf = tdf.drop(['Unnamed: 0'], axis=1)
tdf.head()
```

```python
master_df = pd.merge(wdf, tdf, on='Trip ID')
master_df = master_df.drop(['Unnamed: 0'], axis=1)
master_df
```

```python
master_df = master_df.drop(['Dropoff Census Tract', 'Dropoff Community Area', 'Company', 'Pickup Centroid Location', 'Dropoff Centroid Latitude', 'Dropoff Centroid Longitude', 'Dropoff Centroid Location', 'Community Areas', 'Pickup Census Tract'], axis=1)
master_df
```

```python
newdatestart_list = []

for date in master_df['Trip Start Timestamp']:
    date_obj = datetime.strptime(date, '%m/%d/%Y %H:%M:%S %p')
    new_date = date_obj.strftime('%Y-%m-%dT%H:%M:%S')
    newdatestart_list.append(new_date)

master_df['Trip Start Timestamp'] = newdatestart_list
master_df.head()
```

```python
master_df.columns
```

```python
df1 = master_df.groupby(['icon'], as_index=False).agg({'Trip ID': 'count', 'precipIntensity': 'mean', 'precipProbability': 'mean', 'temperature': 'mean', 'apparentTemperature': 'mean', 'windSpeed': 'mean', 'Trip Seconds': 'mean', 'Trip Miles': 'mean', 'Fare': 'mean', 'Trip Total': 'mean'})
df1
df1.to_csv('./output/groupby_8_icons.csv')
```

```python
df2 = master_df.copy()
df2
```

```python
df2[['day', 'time']] = df2['time'].str.split('T', n=1, expand=True)
df2
df2.to_csv('./output/master_no_grouping.csv')
df2
```

```python
df7['temp range'] = df7['temperature']
df7.to_csv('./output/df7.csv')
df7
```

```python
# temp = int(df7['temp range'])
(df7.loc[df7['apparentTemperature'] > 89, ['temp range']]) = 'High 90+'
(df7.loc[(90 > df7['apparentTemperature']) & (df7['temperature'] > 49), ['temp range']]) = 'Mid 50-89'
(df7.loc[(50 > df7['apparentTemperature']) & (df7['temperature']> 32), ['temp range']]) = 'Low 33-49'
(df7.loc[df7['apparentTemperature'] < 33, ['temp range']]) = 'Freezing'
```

```python
df8 = df7.groupby(['temp range']).agg({'Trip ID': 'count', 'precipIntensity': 'mean', 'precipProbability': 'mean', 'temperature': 'mean', 'apparentTemperature': 'mean', 'windSpeed': 'mean', 'Trip Seconds': 'mean', 'Trip Miles': 'mean', 'Fare': 'mean', 'Trip Total': 'mean'})

df8.to_csv('./output/groupby_high_mid_low_freezing.csv')
df8
```
